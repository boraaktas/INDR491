```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
DATA_FOLDER_PATH = os.getcwd() + '/../data/KS VERI/KS10/'
```

```python
FILE_NAME = 'SIM.KS10 ODASI AYLIK PUE RAPORU.xlsx'
```

```python
PATH = DATA_FOLDER_PATH + FILE_NAME
```

```python
PUE = pd.read_excel(PATH, sheet_name='Sheet1', skiprows=7, usecols='A:B')[:-3]
PUE['Timestamp'] = pd.to_datetime(PUE['Timestamp'], format='%m/%d/%Y %I:%M %p')
PUE.columns = ['Timestamp', 'PUE']
PUE
```

```python
plt.figure(figsize=(20, 10))
plt.plot(PUE['Timestamp'], PUE[PUE.columns[1:]], label=PUE.columns[1:])

plt.xlabel('Timestamp', fontsize=15)
plt.ylabel('PUE', fontsize=15)
plt.title('KS10 PUE', fontsize=20)
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()

plt.show()
```

```python
# save dfs to csv files
PUE.to_csv(os.getcwd() + '/../data/KS VERI/CLEAN_DATA/KS10_PUE.csv', index=False)
```
