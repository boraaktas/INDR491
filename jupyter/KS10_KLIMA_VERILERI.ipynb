```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
DATA_FOLDER_PATH = os.getcwd() + '/../data/KS VERI/KS10/'
```

```python
FILE_NAMES = ['KS10 ODASI KLİMA VERİLERİ 2022 KASIM-ARALIK.xlsx',
              'KS10 ODASI KLİMA VERİLERİ 2023 EYLÜL-EKİM.xlsx',
              'KS10 ODASI KLİMA VERİLERİ 2023 MART-NİSAN.xlsx',
              'KS10 ODASI KLİMA VERİLERİ 2023 MAYIS-HAZİRAN.xlsx',
              'KS10 ODASI KLİMA VERİLERİ 2023 OCAK-ŞUBAT.xlsx',
              'KS10 ODASI KLİMA VERİLERİ 2023 TEMMUZ-AĞUSTOS.xlsx']
```

```python
PATHS = [DATA_FOLDER_PATH + file_name for file_name in FILE_NAMES]
```

```python
ALL_DF = []
for i in range(len(PATHS)):
    print(PATHS[i])
    if i < 4:
        DF = pd.read_excel(PATHS[i], sheet_name='Sheet2', skiprows=7)[:-2]
    else:  
        DF = pd.read_excel(PATHS[i], sheet_name='Sheet2', skiprows=7)[:-3]
    
    new_time_stamps = []
    for timestamp in DF['Timestamp']:
        new_time_stamps.append(pd.to_datetime(timestamp))
        
    DF['Timestamp'] = new_time_stamps
    
    DF = DF.loc[:, ~DF.columns.str.startswith('Unnamed')]
    
    ALL_DF.append(DF)
```

```python
MERGED_DF = pd.concat(ALL_DF, ignore_index=True)
MERGED_DF = MERGED_DF.sort_values(by=['Timestamp'])
MERGED_DF = MERGED_DF.reset_index(drop=True)
MERGED_DF
```

```python
SIC_SET4_DF = pd.read_excel(os.getcwd() + '/../data/KS VERI/KS10/KS10 ODASI KLİMA VERİLERİ 2023 SIC_SET4.xlsx', skiprows=7, usecols='A:B')[:-2]
SIC_SET4_DF['Timestamp'] = pd.to_datetime(list(SIC_SET4_DF['Timestamp']))
SIC_SET4_DF
```

```python
# add the sic set 4 data to the merged dataframe with the same timestamp
MERGED_DF_COPY = MERGED_DF.copy()
MERGED_DF_COPY = MERGED_DF_COPY.merge(SIC_SET4_DF, on='Timestamp', how='left')
MERGED_DF_COPY
```

```python
MERGED_DF = MERGED_DF_COPY.copy()
```

```python
figure, axis = plt.subplots(len(MERGED_DF.columns) - 1, 1, figsize=(20, 70))

figure.subplots_adjust(hspace=1)

for i in range(len(MERGED_DF.columns) - 1):
    axis[i].plot(MERGED_DF['Timestamp'], MERGED_DF[MERGED_DF.columns[i + 1]], label=MERGED_DF.columns[i + 1])
    axis[i].set_title(MERGED_DF.columns[i + 1][10:])
    axis[i].grid()
    axis[i].set_xlabel('Timestamp')

plt.show()
```

```python
# save the dataframe as csv file
MERGED_DF.to_csv(os.getcwd() + '/../data/KS VERI/CLEAN_DATA/KS10_KLIMA_VERILERI.csv', index=False)
```
