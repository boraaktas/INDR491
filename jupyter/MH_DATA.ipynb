```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
DATA_FOLDER_PATH = os.getcwd() + '/../data/KS VERI/'
```

```python
FILE_NAME = 'KS10_FINAL_DATA.csv'
PATH = DATA_FOLDER_PATH + FILE_NAME
```

```python
KS10_REAL_DATA = pd.read_csv(PATH)
KS10_REAL_DATA
```

```python
KS10_MH_DATA = pd.read_csv(DATA_FOLDER_PATH + 'KS10_MH_OUTPUT_DATA_SA.csv')
KS10_MH_DATA
```

```python
MH_PUE = KS10_MH_DATA['PUE']
REAL_PUE = KS10_MH_DATA['KS10_REAL_PUE']
TIMESTAMPS = KS10_MH_DATA['Timestamp']

MH_PUE_REAL_PUE = pd.DataFrame(columns=['Timestamp', 'PUE', 'REAL_PUE'])
MH_PUE_REAL_PUE['Timestamp'] = [None] * 12 + list(TIMESTAMPS)
MH_PUE_REAL_PUE['PUE'] = [None] * 12 + list(MH_PUE)
MH_PUE_REAL_PUE['REAL_PUE'] = [None] * 12 + list(REAL_PUE)
MH_PUE_REAL_PUE
```

```python
# add 12 timestamps before the first timestamp from KS10_REAL_DATA
first_timestamp = KS10_MH_DATA['Timestamp'][0]

OLD_TIMESTAMP_PUE = pd.DataFrame(columns=['Timestamp', 'PUE'])
for i in range(13):
    older_timestamp = pd.to_datetime(first_timestamp) - pd.Timedelta(minutes=5) * i
    older_timestamp_pue = KS10_REAL_DATA[KS10_REAL_DATA['Timestamp'] == str(older_timestamp)]['PUE']
    OLD_TIMESTAMP_PUE.loc[i] = [older_timestamp, older_timestamp_pue.values[0]]
    
OLD_TIMESTAMP_PUE = OLD_TIMESTAMP_PUE.sort_values(by='Timestamp', ascending=True)
OLD_TIMESTAMP_PUE = OLD_TIMESTAMP_PUE.reset_index(drop=True)
OLD_TIMESTAMP_PUE['Timestamp'] = OLD_TIMESTAMP_PUE['Timestamp'].apply(lambda x: str(x))
OLD_TIMESTAMP_PUE
```

```python
ALL_TIMESTAMPS = list(OLD_TIMESTAMP_PUE['Timestamp']) + list(MH_PUE_REAL_PUE['Timestamp'][12:])

# drop duplicated timestamps
ALL_TIMESTAMPS = list(dict.fromkeys(ALL_TIMESTAMPS))
ALL_TIMESTAMPS
```

```python
# plot PUE and REAL_PUE columns
plt.figure(figsize=(13, 6))
plt.plot(OLD_TIMESTAMP_PUE['PUE'], label='OLD SYSTEM PUE')
plt.plot(MH_PUE_REAL_PUE['PUE'], label='OPTIMIZED SYSTEM PUE WITH MH')
plt.plot(MH_PUE_REAL_PUE['REAL_PUE'], label='NORMAL SYSTEM PUE')

# draw vertical line for every 12 data
for i in range(0, len(ALL_TIMESTAMPS), 12):
    plt.axvline(x=i, color='gray', linestyle='--', linewidth=0.5)
    if i == 12:
        plt.axvline(x=i, color='red', linestyle='--', linewidth=3, label='MH START')
# put Timestamp as x-axis for every 12 data
plt.xticks(range(0, len(ALL_TIMESTAMPS), 12), ALL_TIMESTAMPS[::12], rotation=60)
plt.legend()
plt.show()
```
